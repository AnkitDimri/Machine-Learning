```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy.stats import norm
```

```python
x = np.linspace (-10, 10, 100)
y = 2*x + 5
```

```python
def loss (w0, w1, x, y):
    
    l = 0
    for i in range (len (x)):
        
        l += (w0*5 + w1*x [i] - y [i])**2
        
    return l
```

```python
w0 = np.linspace (-10, 10, 200)
w1 = np.linspace (-10, 10, 200)

W0, W1 = np.meshgrid (w0, w1)
```

```python
ax = plt.axes(projection='3d')
fig = plt.figure()
ax.set_xlabel('W0')
ax.set_ylabel('W1')
ax.set_zlabel('Loss incurred')

ax.plot_surface(W0, W1, loss (W0, W1, x, y), rstride=1, cstride=1,cmap='viridis', edgecolor='none')
ax.set_title('surface')
```

```python
def loglikelihood (d, mu, w1, w0):
        
    l = 0
    for x, y in zip (d, mu):
            
        l += np.log (norm.pdf (w1*x + w0, y, 1))
        
    return l
```

```python
ax = plt.axes(projection='3d')
fig = plt.figure()
ax.set_xlabel('W0')
ax.set_ylabel('W1')
ax.set_zlabel('Likelihood')

ax.plot_surface(W0, W1, loglikelihood (x, y, W1, W0), rstride=1, cstride=1,cmap='viridis', edgecolor='none')
ax.set_title('surface')
```

```python
ax = plt.axes(projection='3d')
fig = plt.figure()
ax.set_xlabel('W0')
ax.set_ylabel('W1')

ax.plot_surface(W0, W1, loss (W0, W1, x, y), rstride=1, cstride=1,cmap='viridis', edgecolor='none')
ax.plot_surface(W0, W1, loglikelihood (x, y, W1, W0), rstride=1, cstride=1,cmap='viridis', edgecolor='none')
```
