```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
x = np.linspace (-10, 10, 100)
y = 5*x + 2 + np.random.normal (0,3, 100)
```

```python
plt.scatter (x,y)
z = 5*x+2
plt.plot (x, z, color="red")
plt.show ()
```

```python
# more weight given to intercept or w0 so as to get a greater contribution to error magnitude

def loss (w0, w1, x, y):
    
    l = 0
    for i in range (len (x)):
        
        l += (np.abs(w0*5 + w1*x [i] - y [i]))/1000
        
    return l
```

```python
w0 = np.linspace (-100, 100, 200)
w1 = np.linspace (-100, 100, 200)

W0, W1 = np.meshgrid (w0, w1)
```

```python
ax = plt.axes(projection='3d')
fig = plt.figure()
ax.set_xlabel('W0')
ax.set_ylabel('W1')
ax.set_zlabel('Loss incurred')

ax.plot_surface(W0, W1, loss (W0, W1, x, y), rstride=1, cstride=1,cmap='viridis', edgecolor='none')
ax.set_title('surface')
```

```python
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()

plt.contour (W0, W1, loss (W0, W1, x, y), levels = [i for i in np.arange(-25,25, 1)])
```
