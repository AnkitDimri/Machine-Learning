```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import random
```

```python
def likelihood (d, mu, w1, w0):
        
    l = 1
    for x, y in zip (d, mu):
            
        l *= (norm.pdf (w1*x + w0, y, 1))
        
    return l
```

```python
w0 = np.linspace (-5, 5, 100)
w1 = np.linspace (-5, 5, 100)
```

```python
W0, W1 = np.meshgrid (w0, w1)
```

```python
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()

plt.pcolormesh (W0, W1, norm.pdf (W0) * norm.pdf (W1), cmap=plt.cm.jet)
```

```python
x = []
y = []
x.append (random.uniform (-1,1))
y.append (random.uniform (-1,1))
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()

plt.scatter (x,y)
```

```python
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()

plt.pcolormesh (W0, W1, likelihood (x, y, W1, W0), cmap=plt.cm.jet)
```

```python
prior = norm.pdf (W0) * norm.pdf (W1)
ll = likelihood (x, y, W1, W0)
```

```python
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()
posterior = ll * prior

plt.pcolormesh (W0, W1, posterior, cmap=plt.cm.jet)
```
