```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.stats import norm
import random
```

```python
def likelihood (d, mu, w1, w0):
        
    l = 1
    for x, y in zip (d, mu):
            
        l *= (norm.pdf (w1*x + w0, y, 1))
        
    return l
```

```python
w0 = np.linspace (-3, 3, 200)
w1 = np.linspace (-3, 3, 200)
```

```python
W0, W1 = np.meshgrid (w0, w1)
```

```python
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()
plt.plot(-0.3, 0.5, 'b+', color = "white")
plt.pcolormesh (W0, W1, norm.pdf (W0) * norm.pdf (W1), cmap=plt.cm.jet)
```
