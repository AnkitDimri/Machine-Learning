```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import random
```

```python
def likelihood (d, mu, w1, w0):
        
    l = 1
    for x, y in zip (d, mu):
            
        l *= (norm.pdf (w1*x + w0, y, 1))
        
    return l
```

```python
w0 = np.linspace (-3, 3, 200)
w1 = np.linspace (-3, 3, 200)
```

```python
W0, W1 = np.meshgrid (w0, w1)
```

```python
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()

plt.pcolormesh (W0, W1, norm.pdf (W0) * norm.pdf (W1), cmap=plt.cm.jet)
```

```python
x = []
y = []
xx = np.linspace (-1, 1, 10)
random.shuffle (xx)
yy = 0.5 * xx - 0.5 + np.random.normal (0, 0.5, 10)
r = 0
x.append (xx [r])
y.append (yy [r])
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()

plt.scatter (x,y)
```

```python
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()

plt.pcolormesh (W0, W1, likelihood (x, y, W1, W0), cmap=plt.cm.jet)
```

```python
prior = norm.pdf (W0) * norm.pdf (W1)
ll = likelihood (x, y, W1, W0)
```

```python
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()
posterior = ll * prior

plt.pcolormesh (W0, W1, posterior, cmap=plt.cm.jet)
```

```python
r = r + 1
x.append (xx [r])
y.append (yy [r])
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()

plt.scatter (x,y)
```

```python
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()

plt.xlim (-1,1)
plt.ylim (-1,1)
plt.pcolormesh (W0, W1, likelihood (x, y, W1, W0), cmap=plt.cm.jet)
```

```python
prior = posterior
ll = likelihood (x, y, W1, W0)
```

```python
plt.gca().set_aspect('equal',adjustable='box')
plt.draw()
posterior = ll * prior 
plt.pcolormesh (W0, W1, posterior, cmap=plt.cm.jet)
```
